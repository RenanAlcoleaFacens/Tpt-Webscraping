Abaixo os comandos necessários conforme estudado acima

```python
#Função que retorna os hyperlinks da página de resultados:
#Precisa receber o argumento em BS (Beautiful Soup)
def selects_hyperlinks(pagina_resultadoCVE_BS):

    hyperlink_RefExt=[]

    #Separando a tabela que contém os links:
    hyperlink_table = pagina_resultadoCVE_BS.find('table',class_='table table-striped table-condensed table-bordered detail-table')
    
    #Quantidade de tags que contém links:
    hyperlink_list = hyperlink_table.find_all('a')

    #Executando o comando de separar o link do corpo html através do get_text(), repetindo a qtd de vezes necessária:
    for i in range(0,len(hyperlink_list)):
        hyperlink_RefExt.append(site.find('td',attrs={'data-testid':'vuln-hyperlinks-link-'+str(i)}).get_text())
    
    return hyperlink_RefExt
```

Função que seleciona os links recursiva

```python
def selects_hyperlinks_RECURSIVE(pagina_resultadoCVE_BS,hyperlink_RefExt=0,i=0,counter=0):
    
    if i == 0:
        hyperlink_RefExt=[]
        #Separando a tabela que contém os links e quantidade de tags que contém links:
        counter = len(pagina_resultadoCVE_BS.find('table',class_='table table-striped table-condensed table-bordered detail-table').find_all('a'))

    #Executando o comando de separar o link do corpo html através do get_text(), repetindo a qtd de vezes necessária:    
    hyperlink_RefExt.append(pagina_resultadoCVE_BS.find('td',attrs={'data-testid':'vuln-hyperlinks-link-'+str(i)}).get_text())
    i=i+1

    if i == counter:
        return hyperlink_RefExt
    else:
        return selects_hyperlinks_RECURSIVE(pagina_resultadoCVE_BS,hyperlink_RefExt,i,counter)
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

url = 'https://nvd.nist.gov/vuln/detail/CVE-2022-25157'
#url = 'https://nvd.nist.gov/vuln/detail/CVE-2022-29266'
#url = 'https://nvd.nist.gov/vuln/detail/CVE-2022-29265'

pagina = urlopen(url)
site = BeautifulSoup(pagina,'html.parser')

selects_hyperlinks_RECURSIVE(site)
```
