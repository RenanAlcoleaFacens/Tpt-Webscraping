```python
def selects_score(pagina_resultadoCVE_BS):

    if pagina_resultadoCVE_BS.find(attrs={'data-testid':'vuln-cvss3-panel-score'}):
        NIST = pagina_resultadoCVE_BS.find(attrs={'data-testid':'vuln-cvss3-panel-score'}).get_text()
    else:
        NIST = "N/A"
    
    if pagina_resultadoCVE_BS.find(attrs={'data-testid':'vuln-cvss3-cna-panel-score'}):
        CNA = pagina_resultadoCVE_BS.find(attrs={'data-testid':'vuln-cvss3-cna-panel-score'}).get_text()
    else:
        CNA = "N/A"

    scores = [['NIST Base Score:',NIST],['CNA Base Score:',CNA]]

    return scores
```
